create a list so that we can store the data

```python
df_list = []
```

using a loop to fetch all the data from api

```python
for i in range (1,429):
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=28c6192755d840c0b635b08310602e8e&language=en-US&page={}'.format(i))
    temp_df = pd.DataFrame(response.json()['results'])[['id', 'title','overview','release_date','popularity','vote_average','vote_count']]
    df_list.append(temp_df)
    df = pd.concat(df_list, ignore_index=True)
```
